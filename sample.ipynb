```python
import pandas as pd

# loads Datasets
df = pd.merge(
    # Tokyo
    pd.read_csv("csv/jp_13.csv"),
    # Name
    pd.read_csv("csv/address_name.csv")
)
df
```

```python
def isInside(p, poly):
    if poly:
        inside = False
        oldPoint = poly[-1]
        for newPoint in poly:
            if(newPoint[0] > oldPoint[0]):
                p1 = oldPoint
                p2 = newPoint
            else:
                p1 = newPoint
                p2 = oldPoint

            if ((p1[0] < p[0]) == (p[0] <= p2[0]) and ((p[1] - p1[1]) * (p2[0] - p1[0])) < ((p2[1] - p1[1]) * (p[0] - p1[0]))):
                inside = not inside
                
            oldPoint = newPoint
        return inside
    else:
        return False

# Shibuya-station in Tokyo
point = [139.700, 35.659]
idx_bf = 0
poly = []


for idx, lng, lat, name in df.values:
    if  idx == idx_bf:
        # if same area id
        # makes a figure of area
        poly.append([lng, lat])
    else:
        if isInside(point, poly):
            print(name)
        idx_bf = idx
        poly = []
```
